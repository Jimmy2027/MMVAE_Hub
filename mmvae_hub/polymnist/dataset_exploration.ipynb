```python
from pathlib import Path

from torchvision import transforms

from mmvae_hub.base.utils.flags_utils import get_config_path
from mmvae_hub.base.utils.utils import json2dict
from mmvae_hub.polymnist.PolymnistDataset import PolymnistDataset
from matplotlib import pyplot as plt
```

```python
transform = transforms.Compose([transforms.ToTensor()])
config = json2dict(get_config_path())
test_set = PolymnistDataset(Path(config['dir_data']) / 'train', transform=transform, num_modalities=3)
```

```python
for idx in range(10):
    elem, label = test_set.__getitem__(idx)
    print(label)
    plt.figure(figsize=(10,5))
    plt.subplot(1,3,1)
    plt.imshow(elem['m0'].moveaxis(0,-1))

    plt.subplot(1,3,2)
    plt.imshow(elem['m1'].moveaxis(0,-1))

    plt.subplot(1,3,3)
    plt.imshow(elem['m2'].moveaxis(0,-1))
    plt.show()
```
